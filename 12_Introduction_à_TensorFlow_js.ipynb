# Apprentissage automatique
------------

L'apprentissage automatique (*machine learning en anglais*) se positionne en tant que sous-domaine de l'intelligence artificielle, fournissant des approches mathématiques et statistiques permettant aux ordinateurs d'acquérir des connaissances à partir de données et de résoudre des problèmes sans nécessiter une programmation explicite des tâches.

L'apprentissage automatique [se décompose généralement en deux phases](https://fr.wikipedia.org/wiki/Apprentissage_automatique) :

* **Entraînement :** La première phase implique la création d'un modèle à partir des données, souvent appelées observations. Cette étape dite d'*apprentissage* ou d'*entraînement* est généralement effectuée avant l'utilisation pratique du modèle.

* **Inférence :** Lors de la seconde phase, habituellement lors de la mise en production du système, le modèle est utilisé pour prédire des solutions pour de nouvelles données, distinctes de celles utilisées pour son entraînement. Ces nouvelles données peuvent être soumises pour obtenir le résultat correspondant à la tâche souhaitée.

## Réseaux de neurones

Le concept de [réseau de neurones artificiels (ANN)](https://www.ibm.com/fr-fr/cloud/learn/neural-networks) s'inspire du fonctionnement du cerveau humain, et constitue un sous-ensemble essentiel de l'apprentissage automatique.

Un ANN est composé de différentes couches de neurones artificiels, également appelés nœuds, comprenant : une couche d'entrée, une ou plusieurs couches cachées, et une couche de sortie.

Chaque nœud possède un poids et un seuil associés. Lorsque la sortie d'un nœud dépasse le seuil, le nœud est activé, et il transmet alors des données à la couche suivante.

Lorsqu'un ANN est utilisé pour résoudre un problème, les données d'entrée sont présentées à la couche d'entrée du réseau, généralement sous forme d'un tenseur. La couche de sortie affiche la réponse possible. Pour un problème de classification, par exemple, la réponse peut se présenter sous la forme d'un ensemble de probabilités pour différentes classes possibles.

Le processus supervisé d'entraînement d'un ANN comprend un ensemble d'algorithmes et de techniques visant à présenter des exemples au réseau, pour lesquels les réponses sont connues, afin d'ajuster les poids des neurones constituant le réseau.

Les exemples pour lesquels nous disposons de réponses connues sont appelés **exemples annotés** ou **données étiquetées** (*labeled data* en anglais).

##Types d'apprentissage automatique

Il existe essentiellement quatre catégories d'algorithmes d'apprentissage automatique, chacune apportant une approche distinctive à la résolution de problèmes :

* **Apprentissage supervisé :** Dans cette méthode, des exemples annotés sont utilisés. [Les algorithmes](https://fr.wikipedia.org/wiki/Apprentissage_supervis%C3%A9) cherchent fondamentalement à découvrir ou à apprendre la fonction de prédiction à partir de ces exemples.

* **Apprentissage non supervisé :** [Les algorithmes d'apprentissage non supervisés](https://experiences.microsoft.fr/articles/intelligence-artificielle/apprentissage-supervise-et-non-supervise-quelles-differences/) ne nécessitent pas d'exemples annotés, car les réponses recherchées ne sont pas incluses dans les jeux de données.

* **Apprentissage semi-supervisé :** Cette catégorie de problèmes implique généralement un ensemble de données étiquetées et un ensemble de données non étiquetées.

* **Apprentissage par renforcement :** Dans cette classe de problèmes, le scénario ou l'environnement est souvent modélisé comme un processus de décision de Markov. Un agent (par exemple, un robot) peut alors apprendre les actions à prendre en interagissant avec l'environnement, dans le but d'optimiser une [récompense au fil du temps](https://fr.wikipedia.org/wiki/Apprentissage_par_renforcement).


**L'IA générative**, bien qu'elle ne soit pas explicitement mentionnée parmi les quatre types d'apprentissage automatique, peut être intégrée à certains d'entre eux en fonction de son utilisation. Elle se concentre sur la création de nouvelles données en imitant des modèles appris à partir de données d'entraînement, incluant la génération de textes, d'images, de musique, etc. Selon l'approche, elle peut être liée à l'apprentissage supervisé avec des exemples annotés ou à l'apprentissage non supervisé en explorant des structures latentes pour créer de nouvelles instances. Ainsi, bien que non définie comme une cinquième catégorie, elle peut être vue comme une extension de certains types d'apprentissage automatique existants.

# Introduction à TensorFlow.js
---------

Un cadriciel d'apprentissage automatique fournit essentiellement les outils nécessaires pour construire l'algorithme d'entraînement d'un modèle et pour l'utiliser une fois entraîné. Ce modèle, en quelque sorte, est un algorithme que le développeur n'a pas programmé directement. Il est beaucoup plus petit que l'ensemble de données utilisé pour son entraînement, comprenant les poids du réseau entraîné et la structure du réseau, notamment dans le cas des ANN.

**TensorFlow.js**, développé et entretenu par Google, facilite les tâches d'entraînement, d'inférence et de mise en production des algorithmes d'apprentissage automatique. Il offre un code optimisé pour des ressources telles que les GPU. L'API de TensorFlow.js permet de spécifier la structure ou l'architecture du modèle, de charger les données et de les faire passer par le processus d'apprentissage ou d'inférence.

Les solutions développées peuvent être déployées dans divers environnements :

* Côté application frontale sur tous les navigateurs les plus populaires.

* Côté serveur avec Node.js.

* Sur des appareils mobiles via React Native.

* Applications de bureau sur Electron.

* Applications IoT sur des Raspberry PIs.


Les modèles en TensorFlow.js reposent sur des [graphes orientés acycliques (DAG)](https://fr.wikipedia.org/wiki/Graphe_orient%C3%A9_acyclique), issus du processus d'entraînement et créés avec TensorFlow en Python, puis convertis pour être consommés par TensorFlow.js.

Dans ce cours, nous nous concentrerons sur l'utilisation de modèles existants pour construire quelques applications, en mettant l'accent sur la phase d'inférence. [TensorFlow Hub](https://tfhub.dev/) propose un vaste catalogue de modèles entraînés pour divers types de tâches tels que la classification, la détection d'objets, le traitement du langage naturel, etc.

# Tenseurs
---

Pour entraîner ou utiliser un modèle d'apprentissage automatique, il est crucial de garantir que vos données adoptent un format uniforme, accepté par l'algorithme d'entraînement ou d'inférence.

Lors de la présentation de ces données à un modèle, celui-ci décode et réplique ce qui est en réalité encodé dans la représentation de vos données.

- Fondamentalement, toutes les données doivent être exprimées sous forme numérique pour être comprises par les machines.

- Les tenseurs constituent le choix de prédilection pour représenter les données dans un modèle d'apprentissage automatique.

**Un tenseur est avant tout une structure de données à une ou plusieurs dimensions.** Ils ressemblent à des tableaux JavaScript d'une ou plusieurs dimensions. L'ajout d'une dimension à un tenseur correspond à l'augmentation de son "rang" (*rank* en anglais).

Quelques avantages des tenseurs :

- Les calculs impliquant des tenseurs sont optimisés par TensorFlow.js.

- Des opérations couramment utilisées, telles que la multiplication de matrices, sont exécutées de manière extrêmement rapide grâce à TensorFlow.js.

- Les tenseurs émergent comme un format standard pour représenter des données, même si différentes bibliothèques sont parfois utilisées dans le domaine de l'apprentissage.


## Exemples de manipulations de tenseurs

Nous allons utiliser le module TensorFlow.js sur Node pour les exemples de cette section, mais le même code devrait marcher sur le navigateur.

Exemples de création de tenseurs:

````python
%%javascript --target=node --filename=tensor_creation.mjs

import tf from '@tensorflow/tfjs-node';

const data = [1, 2, 3, 4, 5];

// Création d'un tenseur à partir d'un tableau
const firstTensor = tf.tensor(data);
const secondTensor = tf.tensor1d(data);

// Création d'un tenseur float32
const floatTensor = tf.tensor([1.2, 2.2, 3.3], null, 'float32');

console.log("Nombre de tenseurs:", tf.memory().numTensors);

// Création d'un tenseur int32
const intTensor = tf.tensor([1, 2, 3], null, 'int32');

// Création d'un tenseur booléen
const boolTensor = tf.tensor([true, true, false]);

// Conversion de type de données
const newIntTensor = floatTensor.asType('int32');

console.log("Nombre de tenseurs:", tf.memory().numTensors);

// Contrairement aux tableaux et autres variables JavaScript, les tenseurs
// ne sont pas récupérés par le garbage collector de l'engin JavaScript.
// On doit disposer des tenseurs manuellement.

tf.dispose([firstTensor, secondTensor]);
tf.dispose([floatTensor, intTensor, boolTensor, newIntTensor]);

console.log("Nombre de tenseurs:", tf.memory().numTensors);
````

Pour nettoyer la mémoire automatiquement, nous pouvons mettre les instructions de manipulations de tenseurs dans des blocs `tidy()`:

````python
%%javascript --target=node --filename=tensor_tidy.mjs

import tf from '@tensorflow/tfjs-node';

console.log("Nombre de tenseurs au début :", tf.memory().numTensors);

let maintain;
tf.tidy(() => {
    maintain = tf.tensor([4, 5, 6]);
    const tensor2 = tf.tensor([4, 5, 6]);
    const tensor3 = tf.tensor([4, 5, 6]);
    const returned = tf.tensor([4, 5, 6]);

    console.log("Nombre de tenseurs dans le premier tidy :", tf.memory().numTensors);

    // Protéger un tenseur
    tf.keep(maintain);
});

console.log("Nombre de tenseurs après le premier tidy :", tf.memory().numTensors);

// Disposer des tenseurs manuellement
// Notez que vous ne pouvez pas disposer du tenseur `returned` ici, car tidy l'a déjà fait
maintain.dispose();

console.log("Nombre de tenseurs à la fin du premier tidy :", tf.memory().numTensors);

const result = tf.tidy(() => {
    const x = tf.tensor([1, 2, 3]);
    const y = tf.tensor([4, 5, 6]);

    console.log("Nombre de tenseurs dans le deuxième tidy :", tf.memory().numTensors);

    // Effectue quelques opérations dans tidy
    const sum = x.add(y);

    // La valeur renvoyée pas tidy sera affectée a result
    return sum.square();
});

console.log("Nombre de tenseurs à la fin du deuxième tidy :", tf.memory().numTensors);
result.print();
````

## Des tenseurs d'images

Divers algorithmes d'apprentissage automatique sont conçus pour accomplir des tâches liées aux images. Les tenseurs, en tant que structures de données polyvalentes, jouent un rôle crucial dans la représentation et la manipulation d'images.

Les tenseurs offrent une flexibilité considérable lorsqu'il s'agit de traiter des données visuelles. Ils peuvent être employés pour générer des images, créer des motifs et des remplissages, ainsi que pour augmenter la diversité d'un jeu d'images en altérant les exemples originaux.

Par exemple, grâce à l'utilisation de tenseurs, il devient possible de créer des modèles variés en ajustant les attributs des images, d'appliquer des transformations géométriques, et même de générer des données synthétiques pour enrichir l'ensemble d'apprentissage. La puissance des tenseurs dans la manipulation d'images offre ainsi des perspectives créatives et efficaces dans le domaine de l'apprentissage automatique appliqué à la vision par ordinateur.


## Convertir un tenseur en image (navigateur)

Nous allons examiner quelques exemples illustrant comment charger des images en tant que tenseurs et comment convertir ces tenseurs en images afin de les afficher. Pour visualiser les résultats des opérations dans le navigateur, nous ferons usage d'un élément canvas HTML pour l'affichage des images (via `tf.browser.toPixels`):

````python
%%html

<script src="https://cdn.jsdelivr.net/npm/@tensorflow/tfjs@latest/dist/tf.min.js"></script>

<canvas id="myCanvas"></canvas>

<script>

const canvas = document.getElementById("myCanvas");

// Créer une image avec un contenu aléatoire (tensor 3D de forme [400, 400, 3])
const randomImage = tf.randomUniform([400, 400, 3]);

// Convertir le tenseur en pixels sur le canevas
tf.browser.toPixels(randomImage, canvas).then(() => {
  randomImage.dispose();
});

</script>
````

Un pixel est généralement représenté par trois valeurs correspondant aux couleurs primaires utilisées (Rouge, Vert, Bleu - RGB). Dans un tenseur `int32`, ces valeurs seront comprises entre `0` et `255`. Pour un tenseur de type `float32`, les valeurs des couleurs varieront de `0` à `1`.

La troisième dimension d'un tenseur d'image correspond au nombre de **canaux** (*channels* en anglais). Un tenseur d'image en niveaux de gris peut avoir seulement un canal, tandis qu'un tenseur RGB peut en avoir 3 (pour Rouge, Vert, Bleu) ou 4 (pour Rouge, Vert, Bleu, Alpha dans le cas de RGBA).

## Convertir un tenseur en image (Node)

TensorFlow.js propose plusieurs méthodes dans le module `tf.node` pour créer des images à partir de tenseurs, tandis que pour la manipulation des fichiers d'images sur Node, nous utiliserons le module `node:fs`.

**Remarque :** Il est important de noter que la version de TensorFlow.js destinée au navigateur accepte des tenseurs de type `float32` et `int32`, alors que la version pour Node travaille spécifiquement avec des tenseurs de type `int32`.

````python
%%javascript --target=node --filename=tensor_jpeg.mjs

import * as tf from "@tensorflow/tfjs-node";
import fs from "node:fs/promises";

// Créer une image avec un contenu aléatoire
const tensor = tf.randomUniform([400, 400, 3], 0, 255);

try {
    // Encodage du tenseur au format JPEG
    const encodedImage = await tf.node.encodeJpeg(
        tensor,
        "rgb",   // format
        90,      // quality
        true,    // progressive
        true,    // optimizeSize
        true,    // chromaDownsampling
        "cm",    // densityUnit
        250,     // xDensity
        250,     // yDensity
        "Créé par TFJS Node"
    );

    await fs.writeFile("tensor.jpg", encodedImage);

    console.log("Fichier sauvegardé");
} catch (error) {
    console.error("Une erreur s'est produite :", error);
} finally {
    tensor.dispose();
}
````

## Redimensionner les tenseurs d'image

Parfois le jeu d'images dont nous disposons ne se trouve pas dans le format attendu par un modèle ou algorithme d'apprentissage. Dans ce cas, il faut redimensionner les tenseurs.

TensorFlow.js offre des méthodes pour redimensionner des tenseurs d'images:

* `image.resizeNearestNeighbor` et `image.resizeBilinear`

````python
%%html

<script src="https://cdn.jsdelivr.net/npm/@tensorflow/tfjs@latest/dist/tf.min.js"></script>

<img id="photo" src="https://cdn.glitch.global/efa4fb5c-ccb0-4fd1-8e5e-d9f1d483349d/14424781104_902d42e3d1_c.jpg?v=1699742585962" crossorigin="anonymous"/>
<canvas id="nnCanvas"></canvas>
<canvas id="blCanvas"></canvas>

<script>

  // Taille cible pour le redimensionnement des images
  const newSize = [178, 267];

  const image = document.getElementById("photo");
  const nnCanvas = document.getElementById("nnCanvas");
  const blCanvas = document.getElementById("blCanvas");

  // Charger l'image dans un tenseur
  const imageTensor = tf.browser.fromPixels(image);

  // Redimensionner avec une interpolation Nearest Neighbor
  const nnTensor = tf.image.resizeNearestNeighbor(imageTensor, newSize, true);
  tf.browser.toPixels(nnTensor, nnCanvas).then(() => {
    nnTensor.dispose(); // Libérer la mémoire en disposant du tenseur
  });

  // Redimensionner avec une interpolation Bilinear
  const blTensor = tf.image.resizeBilinear(imageTensor, newSize, true);

  const blTensorInt = blTensor.asType('int32');

  tf.browser.toPixels(blTensorInt, blCanvas).then(() => {
    blTensor.dispose();
    blTensorInt.dispose();
  });

  imageTensor.dispose();

</script>
````

## Lots d'images

En général, les modèles d'apprentissage et les algorithmes d'entraînement traitent des ensembles d'images, également appelés lots d'images.

* De manière similaire à la façon dont un tenseur 1D représente un ensemble de valeurs et qu'un tenseur 2D représente un ensemble de listes de valeurs, les tenseurs 4D peuvent être interprétés comme des lots d'images (c'est-à-dire des lots de tenseurs 3D).

On peut utiliser les méthodes `expandDims` pour augmenter le nombre de dimensions d'un tenseur ou `squeeze` pour effectuer l'opération inverse.

````python
%%html --console

<script src="https://cdn.jsdelivr.net/npm/@tensorflow/tfjs@latest/dist/tf.min.js"></script>

<img id="photo" width="450px"
  src="https://cdn.glitch.global/efa4fb5c-ccb0-4fd1-8e5e-d9f1d483349d/14424781104_902d42e3d1_c.jpg?v=1699742585962"
  crossorigin="anonymous"/>

<canvas id="myCanvas"></canvas>

<script>
  const image = document.getElementById("photo");
  const canvas = document.getElementById("myCanvas");

  const flipped = tf.tidy( () => {
    const tensorImage = tf.expandDims(
      tf.browser.fromPixels(image).asType('float32')
    );

    console.log("Format du tenseur", tensorImage.shape);

    return tf.squeeze(tf.image.flipLeftRight(tensorImage)).asType("int32");
  });

  tf.browser.toPixels(flipped, canvas).then( () => {
    flipped.dispose();
  });

</script>
````

## Modèles SSD (*Single Shot MultiBox Detector*)

Contrairement à certaines architectures de réseau de neurones convolutifs qui nécessitent plusieurs passages à travers le réseau, un modèle SSD (*Single Shot MultiBox Detector*) effectue la détection en un seul passage, ce qui le rend plus rapide et plus efficace.

Le modèle SSD divise l'image d'entrée en plusieurs cellules de différentes tailles et prédit les boîtes englobantes et les classes d'objets associées à chaque cellule. Cette approche permet au modèle de détecter des objets de tailles différentes dans une seule opération.

### MobileNet SSD

MobileNet v2 SSD est un modèle SSD qui utilise MobileNet v2 comme architecture de base. Ce modèle est particulièrement utile dans les applications de détection d'objets en temps réel sur des appareils mobiles ou des systèmes embarqués.Ce modèle détecte les objets définis dans l'ensemble de données COCO, donc capable de détecter 80 classes d'objets.

La sortie du modèle n'est généralement pas utilisable directement en raison du nombre élevé de détections. L'exemple utilise les techniques suivantes pour éliminer les redondances :


- **Topk :** Topk est une opération qui sélectionne les *k* éléments les plus importants d'un ensemble de données. Dans ce contexte, elle est utilisée pour extraire les détections les plus pertinentes après l'opération de détection d'objets.


- **IoU (Intersection over Union - Intersection sur Union) :** L'IoU est une mesure utilisée pour évaluer le degré de chevauchement entre deux zones délimitées, telles que des boîtes englobantes. Il représente le rapport entre la zone d'intersection et celle de l'union des deux zones.


- **NMS (Non-Maximum Suppression) :** La suppression non maximale est une technique employée dans la détection d'objets pour éliminer les détections redondantes en ne conservant que la détection la plus pertinente parmi celles qui se chevauchent.


````python
%%html

<style>
  @import url("https://cdnjs.cloudflare.com/ajax/libs/materialize/1.0.0/css/materialize.min.css");
  .container { margin: 0px; }
  .classify { width: 100%; }
  .img_container { position: relative; }
  .img_container img, canvas { position: absolute; left: 0; }
  section { min-height: 800px; }
</style>

<script src="https://cdn.jsdelivr.net/npm/@tensorflow/tfjs@latest" type="text/javascript"></script>

<header class="container center-align">
    <h3>Exemple d'utilisation du modèle MobileNet SSD</h3>
    <p class="flow-text">Attendez le chargement du modèle. Une fois le modèle chargé,
        l'image sera visible et vous pourrez cliquer dessus pour que
        le modèle puisse trouver les objets.</p>
</header>

<section class="scale-out scale-transition container">
  <div id="image1" class="img_container col s12">
    <img class="classify" alt="Image 1" src="https://cdn.glitch.global/efa4fb5c-ccb0-4fd1-8e5e-d9f1d483349d/guilherme-stecanella-SaVlzqe9068-unsplash.jpg?v=1699913650010"
      crossorigin="anonymous"/>
    <canvas></canvas>
  </div>
  <!-- <div id="image2" class="img_container col s12">
    <img class="classify" alt="Image 2" src="https://cdn.glitch.global/efa4fb5c-ccb0-4fd1-8e5e-d9f1d483349d/amos-bar-zeev-3yzE1SUfbwY-unsplash.jpg?v=1699913649254"
      crossorigin="anonymous"/>
    <canvas></canvas>
  </div>
  <div id="image5" class="img_container col s12">
  <img class="classify" alt="Image 5" src="https://cdn.glitch.global/efa4fb5c-ccb0-4fd1-8e5e-d9f1d483349d/david-tomaseti-Vw2HZQ1FGjU-unsplash.jpg?v=1699913649654"
      crossorigin="anonymous"/>
      <canvas></canvas>
  </div>  -->
</section>

<script type="module">
/**
 * Code adapté de l'exemple du chapitre 6 du livre
 * Learning TensorFlow.js : Powerful Machine Learning in JavaScript
 */
import { COCO_LABELS } from "https://innate-lightning-dracorex.glitch.me/tensorflow/coco/labels.js";

const mainSection = document.querySelector('.scale-out');

class SSDMobilenetClassifier {
    static instance = null;
    static modelURL = 'https://tfhub.dev/tensorflow/tfjs-model/ssd_mobilenet_v2/1/default/1';

    static async getInstance(callback) {
        if (! this.instance ) {
            this.instance = await tf.loadGraphModel(this.modelURL, { fromTFHub: true });
            callback();
        }
        return this.instance;
    }
}

const model = await SSDMobilenetClassifier.getInstance(() => {
    // Pour afficher la section main après le chargement du modèle
    mainSection.classList.add('scale-in');
});

const handleClick = async (event) => {
    const image = event.target;
    const div = image.parentNode;

    // Création d'un lot (batch) de taille 1
    const imageTensor = tf.browser.fromPixels(image);
    const batch = tf.expandDims(imageTensor, 0);

    // Préparation du canvas pour dessiner les boîtes
    const detection = div.querySelector("canvas");
    const ctx = detection.getContext('2d');

    // La taille du canvas doit être la même taille de l'image
    const imgWidth = image.width;
    const imgHeight = image.height;
    detection.width = imgWidth;
    detection.height = imgHeight;
    ctx.font = "16px sans-serif";
    ctx.textBaseline = "top";

    // Configuration des seuils de détection
    const detectionThreshold = 0.4;  // Ignorer les classes dont les probabilités sont inférieures à cette valeur
    const iouThreshold = 0.5;
    const maxBoxes = 20;

    // Exécution de la détection d'objet
    const results = await model.executeAsync(batch);

    // Récupération des valeurs des detections plus pertinentes
    const prominentDetection = tf.topk(results[0]);
    const justBoxes = results[1].squeeze();
    const justValues = prominentDetection.values.squeeze();

    // Déplacer les résultats vers JavaScript de façon concurrente
    const [maxIndices, scores, boxes] = await Promise.all([
        prominentDetection.indices.data(),
        justValues.array(),
        justBoxes.array(),
    ]);

    // Application de la suppression non maximale (NMS) pour éliminer les duplicatas
    const nmsDetections = await tf.image.nonMaxSuppressionWithScoreAsync(
        justBoxes,  // [numBoxes, 4]
        justValues, // [numBoxes]
        maxBoxes,
        iouThreshold,
        detectionThreshold,
        1 // 0 est normal NMS, 1 est Soft-NMS -- https://arxiv.org/pdf/1704.04503.pdf
    );

    // Récupération des indices sélectionnés après NMS
    const chosen = await nmsDetections.selectedIndices.data();

    // Libération de la mémoire utilisée
    tf.dispose([
        results[0],
        results[1],
        nmsDetections.selectedIndices,
        nmsDetections.selectedScores,
        prominentDetection.indices,
        prominentDetection.values,
        imageTensor,
        batch,
        justBoxes,
        justValues,
    ]);

    // Dessin des rectangles autour des objets détectés
    chosen.forEach((detection) => {
        ctx.strokeStyle = "#db0e0e";
        ctx.lineWidth = 4;
        ctx.globalCompositeOperation = "destination-over";
        const detectedIndex = maxIndices[detection];
        const detectedClass = COCO_LABELS[detectedIndex];
        const detectedScore = scores[detection];
        const dBox = boxes[detection];

        // Aucune valeur négative pour les positions de départ
        const startY = dBox[0] > 0 ? dBox[0] * imgHeight : 0;
        const startX = dBox[1] > 0 ? dBox[1] * imgWidth : 0;
        const height = (dBox[2] - dBox[0]) * imgHeight;
        const width = (dBox[3] - dBox[1]) * imgWidth;
        ctx.strokeRect(startX, startY, width, height);

        // Dessiner le fond de l'étiquette
        ctx.globalCompositeOperation = "source-over";
        ctx.fillStyle = "#db0e0e";
        const textHeight = 16;
        const textPad = 6;
        const label = `${detectedClass} ${Math.round(detectedScore * 100)}%`;
        const textWidth = ctx.measureText(label).width;
        ctx.fillRect(startX, startY, textWidth + textPad, textHeight + textPad);

        // Dessiner le texte de l'étiquette en dernier pour qu'il est en haut
        ctx.fillStyle = "white";
        ctx.fillText(label, startX + (textPad / 2), startY + (textPad / 2));
    });
}

const images = mainSection.querySelectorAll('.classify');
for (const img of images) {
    img.addEventListener("click", handleClick);
}
</script>
````

### Le contrôleur question-réponse

Le code du contrôleur Express ci-dessous traite une requête de question-réponse. Il utilise le module `@xenova/transformers` pour créer une instance de pipeline de question-réponse avec un modèle qui utilise une version réduite du transformer BERT. Après la validation des paramètres, le contrôleur exécute le pipeline pour obtenir la réponse à la question et renvoie cette réponse au format JSON.

````python
%%javascript --target=disk --filename=controllers/question-answering.mjs

import { env, pipeline } from '@xenova/transformers';
import Format from 'response-format';
import { body, validationResult } from 'express-validator';

class QnAPipeline {
    static task = 'question-answering';
    static model = 'Xenova/distilbert-base-uncased-distilled-squad';
    static instance = null;

    static async getInstance(progress_callback = null) {
        if (this.instance === null) {
            env.allowRemoteModels = true;
            env.cacheDir = './.cache';
            this.instance = pipeline(this.task, this.model, { progress_callback });
        }
        return this.instance;
    }
}

const questionAnswering = [
    body("context").exists().notEmpty(),
    body("question").exists().notEmpty(),
    async (req, res) => {
        const validation = validationResult(req);
        if (validation.isEmpty()) {
            const {context, question} = req.body;
            const pipeline = await QnAPipeline.getInstance();
            const answer = await pipeline(question, context);
            return res.json(Format.success("OK", answer));
        }
        res.send(Format.badRequest("Contexte et question obligatoires", { errors: validation.array() }));
    }];

export default { questionAnswering  }
````
